```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
N = 100000
Population = np.random.normal(100, size=(N,), scale=3)
Population
```

```python
Population.shape
```

```python
mu = Population.mean()
var = Population.var()
std = Population.std()
mu, var, std
```

```python
sns.distplot(Population, color='blue'); 
```

```python
print(mu, var, std)
samples = []
for n in range(2, 1000, 10):
  samples.append(np.mean(np.random.choice(Population, size=n)))

samples = np.array(samples)
print(samples.mean())
sns.lineplot(x=range(2, 1000, 10), y=samples, color='red');
```

```python
print(mu, var, std)
vars = []
for sms in range(2, 1000, 10):
  vars.append(np.var(np.random.choice(Population, size=sms)))

vars = np.array(vars)
print(vars.mean())
sns.lineplot(x=range(2, 1000, 10), y=vars, color="green");
```

```python
print(mu, var, std)
stds = []
for sms in range(2, 1000, 10):
  stds.append(np.std(np.random.choice(Population, size=sms)))

stds = np.array(stds)
print(stds.mean())
sns.lineplot(x=range(2, 1000, 10), y=stds, color='orange');
```

```python
# variance in RV X' Mean of samples

Omega = [[] for _ in range(2, 100)]
l = 0
for sms in range(2, 100):
  for j in range(100):
    Omega[l].append(np.random.choice(Population, size=sms))
  l+=1

RVX = list(map(lambda x: np.array(list(map(lambda x: x.mean(), x))).var(), Omega))
sns.lineplot(x = range(len(RVX)), y = RVX);
```

Relation between var of sample and var of population is
sample var = population var/n
<br>sample var = 0.09105426701305463
<br>Population var = 9.105116847154948 / 100  

```python
Omega_ = [[] for _ in range(2, 100)]
l = 0
for _ in range(2, 100):
  for j in range(100):
    Omega_[l].append(np.random.choice(Population, size=100))
  l+=1

RVX_ = list(map(lambda x: np.array(list(map(lambda x: x.mean(), x))).var(), Omega_))
print(sum(RVX_)/98)
```
